```python
import matplotlib.pyplot as plt
import graphlab as gl
```

```python
song_data = gl.SFrame('song_data.gl')
```

```python
song_data.head()
```

```python
gl.canvas.set_target('ipynb')
```

```python
song_data['song'].show()
```

```python
len(song_data['user_id'])
```

```python
users = song_data['user_id'].unique()
```

```python
len(users)
```

```python
train_data, test_data = song_data.random_split(.8, seed=0)
```

```python
popularity_model = gl.popularity_recommender.create(train_data, user_id="user_id", item_id = "song")
```

```python
popularity_model.recommend(users=[users[3]])
```

## Build a song recommender with personalization

```python
personalized_model = gl.item_similarity_recommender.create(train_data, user_id='user_id', item_id='song')
```

```python
personalized_model.recommend(users= [users[2]])
```

```python
personalized_model.get_similar_items(['With Or Without You - U2'])
```

```python
%matplotlib inline
model_performance = gl.compare(test_data, [popularity_model, personalized_model], user_sample=0.01)
```

```python
gl.show_comparison(model_performance, [popularity_model, personalized_model])
```
